```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/dataset_indeed-scraper_2024-11-13_11-59-18-322.csv')
```

```python
df
```

```python
# Check for missing values
print(df.isnull().sum())

#Remove rows with missing values
df_dropped = df.dropna()
print(df_dropped)

#Fill missing values with a specific value (e.g., 0)
df_filled_zero = df.fillna(0)
print(df_filled_zero)

#Interpolate missing values
df_interpolated = df.interpolate()
df_interpolated
```

```python
# Drop multiple columns
columns_to_drop = ['description', 'descriptionHTML', 'externalApplyLink','isExpired','jobType','jobType/1','postedAt','postingDateParsed','scrapedAt','url']
df_droppna = df.drop(columns=columns_to_drop)
df_droppna
```

```python
# Drop a single column
df_droppna2 = df_droppna.drop(columns=['urlInput','salary','searchInput/country','searchInput/location'])
df_droppna2
```

```python

# Fill missing values with the mean of the column
# Calculate the mean of the 'salary' column, ignoring NaN values
mean = df_droppna2['rating'].mean()

# Fill NaN values in the 'salary' column with the calculated mean
df_filled_mean = df_droppna2.fillna({'rating': mean})

df_filled_mean
```

```python
# prompt: replacing a null value with the mean

# Method 3: Fill missing values with the mean of the column
# Calculate the mean of the 'rating' column, ignoring NaN values
mean_2 = df_filled_mean['reviewsCount'].mean()

# Fill NaN values in the 'rating' column with the calculated mean
df_filled_2 = df_filled_mean.fillna({'reviewsCount': mean_2})

df_filled_2
```

```python
# Find the most frequent string in the column, excluding NaN values
most_frequent_string = df_filled_2['jobType/0'].mode()[0]

# Replace NaN values with the most frequent string
df_filled_2['jobType/0'] = df_filled_2['jobType/0'].fillna(most_frequent_string)

# Now, df['column_name'] will have the NaN values replaced
# with the most frequent non-null string value in the column.

# Example usage (assuming 'company' is the column you want to fill):

most_frequent_company = df_filled_2['jobType/0'].mode()[0]
df_filled_2['jobType/0'] = df_filled_2['jobType/0'].fillna(most_frequent_company)

df_filled_2
```

```python
# Drop duplicate rows based on all columns
df_no_duplicates = df_filled_2.drop_duplicates()

# Print the DataFrame without duplicates
df_no_duplicates
```

```python
# 1. Histogram of numerical columns
plt.figure(figsize=(10, 6))
df_no_duplicates.hist(bins=10, column=['rating', 'reviewsCount'])  # Customize columns
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
# 2. Bar plot for categorical columns
plt.figure(figsize=(8, 6))
sns.countplot(x='jobType/0', data=df_no_duplicates)  # Customize column
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
plt.title('Count of Job Types')
plt.show()
```
