```python
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data
df = pd.read_csv("user_behavior_dataset.csv")

# Plot the distribution of app usage time
plt.figure(figsize=(10, 6))
sns.histplot(df["App Usage Time (min/day)"])
plt.title("Distribution of App Usage Time")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Frequency")
plt.show()
```

```python
grouped_data = df.groupby(['User Behavior Class', 'Age'])['Battery Drain (mAh/day)'].mean().reset_index()

# Visualization: Battery Drain by User Behavior Class
plt.figure(figsize=(10, 6))
sns.barplot(x='User Behavior Class', y='Battery Drain (mAh/day)', hue='Age', data=grouped_data)
plt.title('Average Battery Drain by User Behavior Class ')
plt.show()

# Visualization: Data Usage by Age Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Data Usage (MB/day)', data=df)  # Make sure 'Data Usage' column exists
plt.title('Data Usage Distribution by Age')
plt.show()

# Correlation analysis
# Assuming 'Battery Drain' column exists (rename if needed)
correlation = df[['Age', 'Battery Drain (mAh/day)', 'Data Usage (MB/day)']].corr()
print(correlation)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Age Distribution by Behavior Class

plt.figure(figsize=(10, 6))
sns.boxplot(x="User Behavior Class", y="Age", data=df)
plt.title("Age Distribution by User Behavior Class")
plt.xlabel("User Behavior Class")
plt.ylabel("Age")
plt.show()
```
